```python
# 7. Write a program which takes 5 inputs from user for different subjectâ€™s marks, total it and generate mark sheet using grades ?

eng = int(input("Enter the marks of english: " ));
math = int(input("Enter the marks of math: " ));
phy = int(input("Enter the marks of phy: " ));
comp = int(input("Enter the marks of comp: " ));
isl = int(input("Enter the marks of isl: " ));
obtain_marks = eng + math + phy + comp + isl;
total_marks = 500;
print("The obtained marks are: ", obtain_marks);
per = (obtain_marks/total_marks)*100;
print("Percentage = ", per);
if per > 100 and per < 0:
    print("Invalid Percentage");
elif per < 100 and per >= 80:
    print("Grade A+");
elif per < 80 and per >= 70:
    print("Grade A");
elif per < 70 and per >= 60:
    print("Grade B");
elif per < 60 and per >= 50:
    print("Grade C");
else:
    print("Fail");
```
